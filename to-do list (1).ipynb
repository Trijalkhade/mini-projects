```python
tasks = []
remove = -1
mark_task = []
def remove():
  view()

  remove = int(input("Enter task number to remove (or 0 to cancel): "))
  if remove == 0:
    print("Removal cancelled.")
    return
  if 0 <= remove < len(tasks):
        removed_task = tasks.pop(remove)
        print(f"Task '{removed_task}' removed.\n")
        return
  elif remove == -1:
        print("Removal cancelled.")
        return
  else:
        print("Invalid task number.")
        return

def add():
  task = input("Enter new task: ")
  priority = input("Enter priority (High, Normal, Low): ").capitalize()
  tasks.append(task)
  print(f"Task '{task}' added.\n")

def view():

  if not tasks:
    print("No tasks in the list.")
    return

  for i, task in enumerate(tasks):
    if i in mark_task:
      print(f"{i+1}. [X] {task}")
    else:
      print(f"{i+1}. {task}")

def mark():
  view()

  num = int(input("Enter task number to mark complete (or 0 to cancel): ")) - 1
  mark_task.append(num)

def main():

  print("\nChoose an action:")
  print("1. Add task")
  print("2. View tasks")
  print("3. Remove task")
  print("4. Mark task complete")
  print("5. Save and Exit")
  while True:
    choice = input("Enter your choice: ")
    if choice == "1":
      add()
    elif choice == "2":
      view()
    elif choice == "3":
      remove()
    elif choice == "4":
      mark()
    elif choice == "5":
      print("Tasks Exited successfully")
      break
    else:
      print("Invalid choice. Please try again.")

    if not 5:
      break
if __name__ == "__main__":
  main()


```
