```python
class Item:
  def __init__(self,name: str,price: float,quantity: int=1 ):
    self.price = float(price)
    self.name = str(name)
    self.quantity = int(quantity)
```

```python
class ShoppingCart:
  def __init__(self):
    self.items = []
  def add(self,item):
    self.items.append(item)
  def total(self):
    return sum((item.price*item.quantity) for item in self.items)
  def __len__(self):
    return len(self.items)
  def __iter__(self):
    return iter(self.items)
```

```python
# The script can be imported and it wont run the below code
if __name__ == '__main__':
  item = Item("car",649000)
  item1 = Item("bike",129000,2)
  cart = ShoppingCart()
  cart.add(item)
  cart.add(item1)
  for item in cart:
    print(item.name,item.price,item.quantity)
  print("Total: ",cart.total())
  print("Distinct items in cart: ",len(cart))
```
