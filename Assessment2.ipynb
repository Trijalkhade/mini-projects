# **EVEN / ODD**

```python
n = int(input())
```

### **1. list**

```python
even_odd = ["Even","Odd"]
print(even_odd[n%2])
```

### **2. slicing**

```python
print("EOvdedn"[n%2::2])
```

### **3. conditional**

```python
if n%2 == 0:
  print("Even")
else:
  print("Odd")
```

### **4. string comprehension**

```python
print("Even" if n%2==0 else "Odd")
```

# **Visualizing Time Complexity**

```python
# import libraries for plotting data
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# Define lists and necessary data for visual representation
if_else_time_list = []
ternary_time_list = []
list_time_list = []
slicing_time_list = []
length_list = [num for num in range(10,1001,100)]
methods = ["If-else block","Ternary Operator","List-based Approach","String Slicing Approach"]
```

```python
import time

# Function to measure time for different lengths of numbers
def measure_time(n):
    # Code 1: If-else block
    start_time = time.time()
    if n % 2 == 0:
        pass
    else:
        pass
    if_else_time = time.time() - start_time

    # Code 2: Ternary Operator
    start_time = time.time()
    ("Even" if n % 2 == 0 else "Odd")
    ternary_time = time.time() - start_time

    # Code 3: List-based Approach
    start_time = time.time()
    even_odd = ["Even", "Odd"]
    even_odd[n % 2]
    list_time = time.time() - start_time

    # Code 4: String Slicing Approach
    start_time = time.time()
    "EOvdedn"[n % 2::2]
    slicing_time = time.time() - start_time

    return if_else_time, ternary_time, list_time, slicing_time

# Test with increasing number of digits
for length in range(10, 1001, 100):  # From 10 digits to 1000 digits
    n = int("1" * length)  # Create a number with 'length' digits
    times = measure_time(n)
    print(f"Number length: {length} digits")
    print(f"Time taken by If-Else block: {times[0]} seconds")
    if_else_time_list.append(times[0])
    print(f"Time taken by Ternary Operator: {times[1]} seconds")
    ternary_time_list.append(times[1])
    print(f"Time taken by List-based Approach: {times[2]} seconds")
    list_time_list.append(times[2])
    print(f"Time taken by String Slicing Approach: {times[3]} seconds")
    slicing_time_list.append(times[3])
    print("=" * 50)
```

```python
# Create a DataFrame from the data
data = {
    'Length': length_list,
    'If-else block': if_else_time_list,
    'Ternary Operator': ternary_time_list,
    'List-based Approach': list_time_list,
    'String Slicing Approach': slicing_time_list
}

df = pd.DataFrame(data)
```

```python
df
```

```python
plt.figure(figsize=(10,6),dpi=100,facecolor="white")
# Combine the plots by same X-axis,Y-axis and Data provided
sns.lineplot(x="Length", y="If-else block", data=df, label="If-else block", color="blue", marker="o")
sns.lineplot(x="Length", y="Ternary Operator", data=df, label="Ternary Operator", color="green", marker="o")
sns.lineplot(x="Length", y="List-based Approach", data=df, label="List-based Approach", color="orange", marker="o")
sns.lineplot(x="Length", y="String Slicing Approach", data=df, label="String Slicing Approach", color="red", marker="o")

# Layout of the Barplot
plt.title("Time Taken by Various methods over the range of input size", fontsize=16, fontweight="bold")
plt.xlabel("Number Length (Digits)", fontsize=12)
plt.ylabel("Time (Seconds)", fontsize=12)
plt.legend(title="Methods")
plt.show()
```

```python
# Find avg time required
if_else_avg = df["If-else block"].mean()
ternary_avg = df["Ternary Operator"].mean()
list_avg = df["List-based Approach"].mean()
slicing_avg = df["String Slicing Approach"].mean()
```

```python
# Ranking Based on Mean values
rank = {"If-else block":if_else_avg,
        "Ternary Operator": ternary_avg,
        "List-based Approach": list_avg,
        "String Slicing Approach": slicing_avg
        }
sorted_rank = (sorted(rank.items(), key=lambda item: item[1], reverse=False))

print("Best to worst:")
for name,val in sorted_rank:
  print(f"{name} : {val}")
```
